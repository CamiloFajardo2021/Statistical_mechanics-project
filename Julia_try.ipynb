```julia
using Plots 
using Printf

function fz1(z, kappa)
    a = (1 - sqrt(kappa/(16+kappa)))/2
    return ((z-a).^a).*(z-(a-1)).^(1-a)
end

function fz2(z, kappa)
    a = (1 - sqrt(kappa/(16+kappa)))/2
    return ((z+a).^a).*(z-(1-a)).^(1-a)
end

#SLE parameters
n = 10000 #Total number of points
kappa = 2 #fractal parameter kappa
#SLE evolution
f1 = zeros(Complex{Float64}, n) #Initialization of the curve
for h in 2:n
    p = rand(1)[1]
    if p>0.5
       f1[h] =  fz1(f1[h-1],kappa) #recursive call of the first map if p>0.5
    else
       f1[h] =  fz2(f1[h-1],kappa) #recursive call of the first map if p>0.5
    end
end

x = real.(f1)
y = imag.(f1)

print(size(f1))


p = plot(x, y, size = (400,500))
display(p)
```
