```python
import numpy as np
import matplotlib.pyplot as plt

C = np.zeros(100,dtype=np.complex_)
x = np.linspace(0, 1, len(C))
y = np.exp(x)

for i in range(len(C)):
    C[i] = x[i]+(x[i])**(0.5)*1.j 
    y[i] = C[i].imag

#Funtion solution of the SLE transformation equation
g = lambda z: z.real + np.sqrt((z-z.real)**2 + 4*z.imag**2) 

#implementing the transformation
t = np.zeros(len(C)-1)
transf = np.zeros(len(t))
for n in range(len(t)):
    transf[n] = g(C[n+1])

plt.figure(figsize=(10,10))
plt.plot(x,y, color = 'k')
plt.scatter(transf, np.zeros(len(transf)), color = 'g')
# plt.ylim(0,15)
```
